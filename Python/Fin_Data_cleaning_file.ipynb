```python
# Fix — add the two missing counties
mandera_turkana = pd.DataFrame([
    {
        "county_name"       : "Mandera",
        "county_id"         : 9,
        "region"            : "North Eastern",
        "active_agents_2024": 1900,
        "population_2024"   : 867000,
        "settlement_type"   : "Rural",
    },
    {
        "county_name"       : "Turkana",
        "county_id"         : 23,
        "region"            : "Rift Valley",
        "active_agents_2024": 2600,
        "population_2024"   : 926000,
        "settlement_type"   : "Rural",
    },
])


agent_network = pd.concat([agent_network, mandera_turkana], ignore_index=True)
agent_network = agent_network.sort_values("county_id").reset_index(drop=True)
```

```python
# Fix 1 — correct formula: agents / population × 1,000
agent_network["agents_per_10k_pop"] = (
    agent_network["active_agents_2024"] / agent_network["population_2024"] * 1000
).round(2)
```

```python
# Also recompute agents_per_1000_adults (adults ≈ 55% of population)
agent_network["agents_per_1000_adults"] = (
    agent_network["active_agents_2024"] / (agent_network["population_2024"] * 0.55) * 1000
).round(2)
```

```python
print(agent_network.head())
```

```python
# Fix 2 — rederive tier from corrected density values
def density_tier(x):
    if x >= 10:  return "High"
    elif x >= 6: return "Medium"
    else:        return "Low"
 
agent_network["agent_density_tier"] = agent_network["agents_per_10k_pop"].apply(density_tier)
```

```python
# Remove duplicates
agent_network= agent_network.drop_duplicates(subset= ['county_name', 'county_id'])
```

```python
# Fix — rename % columns
annual_kpis = annual_kpis.rename(columns={
    "yoy_value_growth_%"  : "yoy_value_growth_pct",
    "yoy_volume_growth_%" : "yoy_volume_growth_pct",
    "formal_inclusion_%"  : "formal_inclusion_pct",
    "mobile_money_gdp_%"  : "mobile_money_gdp_pct",
})
annual_kpis= pd.DataFrame(annual_kpis)
```

```python
annual_kpis["yoy_value_growth_pct"]  = annual_kpis["yoy_value_growth_pct"].fillna(0.0)
annual_kpis["yoy_volume_growth_pct"] = annual_kpis["yoy_volume_growth_pct"].fillna(0.0)
print(annual_kpis[["yoy_value_growth_pct","yoy_volume_growth_pct"]])
```

```python
# Fix — interpolate the 7 missing years
# formal_inclusion_pct only has values for survey years: 2019, 2021, 2024
# Linear interpolation fills the gaps for trend charts
# The original NaN values are preserved; a new column holds the filled version
annual_kpis = annual_kpis.sort_values("year").reset_index(drop=True)
 
annual_kpis["formal_inclusion_pct_interp"] = (
    annual_kpis["formal_inclusion_pct"]
    .interpolate(method="linear", limit_direction="both")
    .round(4)
)
```
